## Zone files:

A zone file (may) contain:

- **TTL** defualt time to live
- **SOA** STart of Authority (needed)
- **NS** record (Name server, needed)
- **MX** record (Mail servers, not needed)
- **A** record (A record, needed)
- **CNAME** record

We are going to need a **SOA** record as well as an **A** record. In addition w must provide a **NS** record for our nameserver.

### **SOA** Record (Start of Authority):

- **Name:** Name of the domain
- **TTL**: We could set a different TTL value for each domain (not needed) 
- **Class:** The default path used for all resource records are **IN**, which stands for internet.
- **RR**: This specifies the resource record name, which is **SOA** here.
- **Nameserver**: This is the primary name server for this domain/zone. *Please note the fact that there must be a A record for this name server later in the same zone file that will specify the IP address of that name server. So a second entry needs to be added to the zone file*
- **Email**: Administrative contact (not needed here)
- **Serialnumber**: Its one of the important entries inside the SOA record in a zone file. Serial number tells the modification date of the zone file. Every time you modify the zone file, you also need to modify the serial number. Serial numbers in zone file follows a date format of **yyyymmddss**. Which means if you have edited your zone file on 12th September 2013, your serial number will be **2013091200**. 
- **Refresh**:  Indicates the time after which the secondary or slave DNS server for this domain re-fetches the SOA record for this zone.
- **Retry**: Specifies the retry interval if the slave will take, in case of a failure.
- **Expiry**: This specifies the duration after which the slave name server will stop responding to DNS queries if the connection to master server cannot be established by following retry interval.
- **MIN**: Indicates the duration for which caching name servers (like your local name server) will cache the domain value

We will define our **SOA** entry as follows:

**kiener.sec21.tech. 3600 IN SOA ns.kiener.sec21.tech. hostmaster.invalid. 2021102700 600 600 604800 60**

That is "*domainname TTL Class RR-name name-of-nameserver email refresh retry expiry min*"

### NS record

This record specifies the DNS server responsible for this domain.
The IP address for the name server needs to be defined int the zone file as an **A** record.

We will define our NS record as:

**kiener.sec21.tech. 3600 IN NS ns.kiener.sec21.tech.**

That is "*domainname TTL Class RR-name name-of-nameserver*"

### A record

A record or address record is used to add IP address for a hostname in a zone. It is the highly used resource record in a zone file.

We will define our A record as:

**ns.kiener.sec21.tech. 3600 IN A 192.42.43.42**

That is "*domainname TTL Class RR-name public IPv4-address*"
